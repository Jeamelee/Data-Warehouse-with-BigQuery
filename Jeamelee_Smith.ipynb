```python
# Function to generate a transaction number
def generate_transaction_number(n):
    return ['TXN' + str(i).zfill(8) for i in range(1, n + 1)]

# Generate the sale quantity and prices
def generate_sales_data(n):
    sale_quantity = np.random.randint(1, 10, size=n)
    regular_unit_price = np.round(np.random.uniform(5.99, 199.99, size=n), 2)
    # 25% of the transactions discounted
    discounts = np.random.choice([0, 0.1, 0.15, 0.2, 0.25], size=n, p=[0.75, 0.05, 0.05, 0.05, 0.1])
    discount_unit_price = np.round(regular_unit_price * discounts, 2)
    net_unit_price = regular_unit_price - discount_unit_price
    return sale_quantity, regular_unit_price, discount_unit_price, net_unit_price

num_records = 4000  # The number of sales transactions to generate

# Generate transaction numbers
transaction_numbers = generate_transaction_number(num_records)
```
